### Perform DMD

$\frac{\partial}{\partial t}\mathbf{x} (t) = \mathbf L_0 \mathbf{x} (t)$

$\mathbf{X} := \begin{bmatrix} \mathbf{x}_1 \mathbf{x}_2 \dots \mathbf{x}_{M-1}\end{bmatrix}$

$\mathbf{X}' := \begin{bmatrix} \mathbf{x}_2 \mathbf{x}_3 \dots \mathbf{x}_M \end{bmatrix}$

$ \mathbf{X}' = A \mathbf{X}$

$  A = \mathbf{X}' \operatorname{pinv} ( \mathbf{X} )$

```julia
function DMD_A(Xᵍ, Xᵉ, Xˣ, Xʸ)
    
    X⁻=  hcat( Xᵍ[:,1:end-1], Xᵉ[:,1:end-1], Xˣ[:,1:end-1], Xʸ[:,1:end-1]) 
    X⁺ = hcat( Xᵍ[:,2:end],   Xᵉ[:,2:end],   Xˣ[:,2:end], Xʸ[:,2:end]) 
    
    return   X⁺ * pinv(X⁻)
    
end
```

```julia
function propagate_DMD_A(A, X₀, m) 
    Xˢⁱᵈ = [X₀]
    for i in 2:m
        push!(Xˢⁱᵈ, A*Xˢⁱᵈ[end])
    end
    return hcat(Xˢⁱᵈ...)
end
```

```julia
function DMD_SVD(Y, r, Δt)
    
    X₋ = Y[:,1:end-1]
    X₊ = Y[:,2:end]
    U, Σ, V = svd(X₋)
    
    Uʳ = U[:, 1:r] #12 x 4
    
    Σʳ = diagm(Σ[1:r])
    Vʳ = V[:, 1:r]
    
    Ã = Uʳ' * X₊ * Vʳ / Σʳ
    
    Λ, W = eigen(Ã)

    Φ = X₊ * Vʳ / Σʳ * W

    Ω = log.(Λ)/Δt

    x₁ = X₋[:,1]

    b₁ = Φ \ x₁
    
    return Φ, Ω, b₁

end    
```

```julia
function DMD_reconstruct(Φ, Ω, b₁, Δt, steps)
    
    Yᵈᵐᵈ = hcat([real.(Φ * (b₁ .* exp.(Ω * (i * Δt)))) for i in 0:steps]...)
    
    return Yᵈᵐᵈ

end
```

```julia
Φ, Ω, b₁ = DMD_SVD(bˣₐₙ, 3, Δtₑₓ)
```

```julia
steps = size(bˣₐₙ,2)-1

bˣᵈᵐᵈ = DMD_reconstruct(Φ, Ω, b₁, Δtₑₓ, steps)

norm(bˣᵈᵐᵈ - bˣₐₙ)
```

```julia
maxf = []
minf = []
avgf = []
stdf = []

Y¹² = vcat(bᵍₐₙ, bᵉₐₙ, bˣₐₙ, bʸₐₙ)
steps = size(Y¹²,2)-1

for r in 1:11
       
    Φ, Ω, b₁ = DMD_SVD(Y¹², r, Δtₑₓ)

    Y¹²ᵈᵐᵈ = DMD_reconstruct(Φ, Ω, b₁, Δtₑₓ, steps)

    bᵍᵈᵐᵈ = Y¹²ᵈᵐᵈ[1:3,:]
    bᵉᵈᵐᵈ = Y¹²ᵈᵐᵈ[4:6,:]
    bˣᵈᵐᵈ = Y¹²ᵈᵐᵈ[7:9,:]
    bʸᵈᵐᵈ = Y¹²ᵈᵐᵈ[10:12,:]

    ρᵍᵈᵐᵈ = LiPoSID.rho_series_from_bloch(bᵍᵈᵐᵈ)
    ρᵉᵈᵐᵈ = LiPoSID.rho_series_from_bloch(bᵉᵈᵐᵈ)
    ρˣᵈᵐᵈ = LiPoSID.rho_series_from_bloch(bˣᵈᵐᵈ)
    ρʸᵈᵐᵈ = LiPoSID.rho_series_from_bloch(bʸᵈᵐᵈ)

    fᵍᵈᵐᵈ = LiPoSID.fidelity_series(basis, ρᵍₐₙ, ρᵍᵈᵐᵈ)
    fᵉᵈᵐᵈ = LiPoSID.fidelity_series(basis, ρᵉₐₙ, ρᵉᵈᵐᵈ)
    fˣᵈᵐᵈ = LiPoSID.fidelity_series(basis, ρˣₐₙ, ρˣᵈᵐᵈ)
    fʸᵈᵐᵈ = LiPoSID.fidelity_series(basis, ρʸₐₙ, ρʸᵈᵐᵈ)
    
    push!(minf, minimum([minimum(fᵍᵈᵐᵈ), minimum(fᵉᵈᵐᵈ), minimum(fˣᵈᵐᵈ), minimum(fʸᵈᵐᵈ)]))
    
    push!(maxf, maximum([maximum(fᵍᵈᵐᵈ), maximum(fᵉᵈᵐᵈ), maximum(fˣᵈᵐᵈ), maximum(fʸᵈᵐᵈ)]))
    
    push!(avgf, mean([mean(fᵍᵈᵐᵈ), mean(fᵉᵈᵐᵈ), mean(fˣᵈᵐᵈ), mean(fʸᵈᵐᵈ)]))
        
    push!(stdf, std([std(fᵍᵈᵐᵈ), std(fᵉᵈᵐᵈ), std(fˣᵈᵐᵈ), std(fʸᵈᵐᵈ)]))
    
end  
```

```julia
plot(minf, label = "min")
plot!(maxf, label = "max")
plot!(avgf, label = "mean")
plot!(avgf - stdf, label = "mean - std")
plot!(avgf + stdf, label = "mean + std")
```
